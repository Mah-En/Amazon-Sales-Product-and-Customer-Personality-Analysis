# Part 1: Amazon Sales Analysis

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/amazon-dataset/amazon.csv')
df.head()
```

### Data Preproccess

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False)
df['discounted_price'] = df['discounted_price'].str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['discounted_price'] = df['discounted_price'].fillna(df['discounted_price'].median()) # No need

df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False)
df['actual_price'] = df['actual_price'].str.replace(',', '', regex=False)
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['actual_price'] = df['actual_price'].fillna(df['actual_price'].median())             # No need

df['rating'] = df['rating'].str.split('.')
df = df.explode('rating')
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df.dropna(subset=['rating'], inplace=True)                                              # No need

df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].median())
```

```python
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
df['rating_category'] = df['rating'].apply(lambda x: 'High' if x >= 4 else 'Low')

label_encoder = LabelEncoder()
df['category_encoded'] = label_encoder.fit_transform(df['category'])

scaler = MinMaxScaler()
df[['discounted_price', 'actual_price']] = scaler.fit_transform(df[['discounted_price', 'actual_price']])
```

## Statistical Questions

### 1. Does the discounted price significantly impact product ratings?

```python
import scipy.stats as stats
```

```python
correlation, p_val = stats.spearmanr(df['discounted_price'], df['rating'])

print(f"Spearman Correlation Coefficient: {correlation}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

P-values range from −1 to +1.\
p=+1 → Perfect positive monotonic relationship.\
p=−1 → Perfect negative monotonic relationship.\
p=0 → No monotonic relationship.


The 0.05 threshold is a standard balance between avoiding false positives and false negatives.\
Using 0.5 as a threshold would be far too lenient, leading to a high rate of false positives and unreliable conclusions.

### 2. Are product categories and high/low ratings independent?

```python
import numpy as np
from scipy.stats import chi2_contingency
```

```python
df['rating_category'] = np.where(df['rating'] >= 4, 'High', 'Low')

contingency_table = pd.crosstab(df['category'], df['rating_category'])

print("Contingency Table:")
contingency_table
```

```python
chi2, p_val, dof, expected = chi2_contingency(contingency_table)

print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

if p_val < 0.05:
    print("\n Reject the null hypothesis.\n A significant relationship between product categories and ratings.")
else:
    print("\n Fail to reject the null hypothesis.\n No significant relationship between product categories and ratings.")
```

Since the p-value is very close to 1, The two variables are likely independent.

### 3. Is there a significant difference in ratings between high-discount and low-discount products?

```python
from scipy.stats import ttest_ind, levene
```

```python
threshold = 50  
high_discount = df[df['discount_percentage'] > threshold]['rating']
low_discount = df[df['discount_percentage'] <= threshold]['rating']

levene_test = levene(high_discount, low_discount)
print(f"Levene's Test p-value: {levene_test.pvalue}")

if levene_test.pvalue > 0.05:
    t_stat, p_val = ttest_ind(high_discount, low_discount, equal_var=True)
else:
    t_stat, p_val = ttest_ind(high_discount, low_discount, equal_var=False)
```

```python
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis.\n A significant difference in ratings between high-discount and low-discount products.")
else:
    print("Fail to reject the null hypothesis.\n No significant difference in ratings between high-discount and low-discount products.")
```

The t-statistic tells us how many standard errors the difference between the means is away from zero.

A larger absolute value of the t-statistic indicates a greater difference between the groups.

### 4. Do more expensive products receive higher ratings on average?

```python
from scipy.stats import f_oneway, levene

low_threshold = df['actual_price'].quantile(0.33)
high_threshold = df['actual_price'].quantile(0.66)

df['price_group'] = pd.cut(
    df['actual_price'],
    bins=[-np.inf, low_threshold, high_threshold, np.inf],
    labels=['Low', 'Medium', 'High']
)

levene_test = levene(
    df[df['price_group'] == 'Low']['rating'],
    df[df['price_group'] == 'Medium']['rating'],
    df[df['price_group'] == 'High']['rating']
)
print(f"Levene's Test p-value: {levene_test.pvalue}")

f_stat, p_val = f_oneway(
    df[df['price_group'] == 'Low']['rating'],
    df[df['price_group'] == 'Medium']['rating'],
    df[df['price_group'] == 'High']['rating']
)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis. A significant difference in ratings across price groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in ratings across price groups.")
```

```python
import numpy as np
from scipy.stats import kstest, shapiro, norm
```

#### Kolmogorov-Smirnov Test

```python
mean, std = np.mean(rating_counts), np.std(rating_counts)
ks_stat, ks_p_val = kstest(rating_counts, 'norm', args=(mean, std))
```

#### Shapiro-Wilk Test

```python
shapiro_stat, shapiro_p_val = shapiro(rating_counts)
```

```python
print("Kolmogorov-Smirnov Test:")
print(f"KS Statistic: {ks_stat}")
print(f"P-value: {ks_p_val}")

print("\nShapiro-Wilk Test:")
print(f"Shapiro Statistic: {shapiro_stat}")
print(f"P-value: {shapiro_p_val}")

alpha = 0.05

if ks_p_val < alpha:
    print("\nKolmogorov-Smirnov Test: Reject the null hypothesis. The data is not normally distributed.")
else:
    print("\nKolmogorov-Smirnov Test: Fail to reject the null hypothesis. The data is normally distributed.")

if shapiro_p_val < alpha:
    print("Shapiro-Wilk Test: Reject the null hypothesis. The data is not normally distributed.")
else:
    print("Shapiro-Wilk Test: Fail to reject the null hypothesis. The data is normally distributed.")
```

The KS statistic measures the maximum difference between the empirical distribution of the data and the theoretical normal distribution.

The Shapiro-Wilk Test is more sensitive to deviations from normality, especially for small to moderate sample sizes.

A p-value less than the significance level (α=0.05) indicates that the data significantly deviates from a normal distribution.

Conclusion: Reject the null hypothesis. The data is not normally distributed.

### 2. Does the marketing campaign influence spending behavior across customer groups?

```python
from scipy.stats import shapiro
```

```python
df = df.copy()
df.loc[:, 'AcceptedCmp'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].max(axis=1)
df.loc[:, 'TotalSpending'] = (
    df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
)
camp = df.loc[df['AcceptedCmp'] == 1, 'TotalSpending']
no_camp = df.loc[df['AcceptedCmp'] == 0, 'TotalSpending']
```

```python
stat, p_camp = shapiro(camp)
stat, no_camp = shapiro(no_camp)

print(f"Campaign Group: p-value = {p_camp}")
print(f"No-Campaign Group: p-value = {no_camp}")

if p_camp >= 0.05 and no_camp >= 0.05:
    from scipy.stats import ttest_ind
    t_stat, p_val = ttest_ind(camp, no_camp)
    print(f"T-statistic: {t_stat}, p-value: {p_val}")
else:
    from scipy.stats import mannwhitneyu
    u_stat, p_val = mannwhitneyu(camp, no_camp)
    print(f"U-statistic: {u_stat}, p-value: {p_val}")


if p_val < 0.05:
    print("Reject the null hypothesis: The marketing campaign significantly influences spending behavior.")
else:
    print("Fail to reject the null hypothesis: The marketing campaign does not significantly influence spending behavior.")
```

Customers who received the marketing campaign have significantly different spending behavior compared to those who did not.

The extremely small p-value indicates that this result is highly statistically significant.

```python
stat, p_children = shapiro(child)
stat, p_no_children = shapiro(no_child)

print(f"children Group: p-value = {p_children}")
print(f"No-children Group: p-value = {p_no_children}")

u_stat, p_val = mannwhitneyu(child, no_child)

print(f"U-statistic: {u_stat}, p-value: {p_val}")


if p_val < 0.05:
    print("Reject the null hypothesis: Customers with children spend differently than those without children.")
else:
    print("Fail to reject the null hypothesis: Customers with children do not spend differently than those without children.")
```

Customers without children tend to spend more than customers with children.

The extremely small p-value indicates that this result is highly statistically significant.

### 5. Is there a relationship between customer education level and acceptance of promotional campaigns?

```python
df.loc[:, 'AnyCampaign'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].max(axis=1)

contingency_table = pd.crosstab(df['Education'], df['AnyCampaign'])

print(contingency_table)
```

```python
from scipy.stats import chi2_contingency

chi2_statistic, p_val, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Square Statistic: {chi2_statistic}, p-value: {p_val}")

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis: There is a significant association between education level and campaign acceptance.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between education level and campaign acceptance.")
```

The p-value (0.175) is greater than 0.05, which means we do not have sufficient evidence to reject the null hypothesis.

This suggests that education level does not significantly influence whether customers accept promotional campaigns.

In other words, customers across different education levels are equally likely (or unlikely) to accept promotional campaigns.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Education', y='AnyCampaign', data=df, errorbar=None)  # Use errorbar=None instead of ci=None
plt.title('Campaign Acceptance by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion of Campaign Acceptance')
plt.show()
```
